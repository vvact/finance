```python
import requests
import pandas as pd

def get_binance_data(symbol="BTCUSDT", interval="1h", limit=500):
    url = f"https://api.binance.com/api/v3/klines?symbol={symbol}&interval={interval}&limit={limit}"
    response = requests.get(url).json()

    df = pd.DataFrame(response, columns=["timestamp", "open", "high", "low", "close", "volume", 
                                         "close_time", "qav", "num_trades", "taker_base_vol", "taker_quote_vol", "ignore"])

    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="ms") + pd.Timedelta(hours=3)
    df[["open", "high", "low", "close", "volume", "taker_base_vol", "taker_quote_vol"]] = df[["open", "high", "low", "close", "volume", "taker_base_vol", "taker_quote_vol"]].astype(float)

    return df[["timestamp", "open", "high", "low", "close", "volume", "taker_base_vol", "taker_quote_vol"]]

btc_df = get_binance_data()
display(btc_df)
```

```python
import pandas as pd
import numpy as np

def add_features(df):
    # Simple Moving Averages (SMA)
    df["SMA_10"] = df["close"].rolling(window=10).mean()
    df["SMA_50"] = df["close"].rolling(window=50).mean()
    
    # Exponential Moving Averages (EMA)
    df["EMA_10"] = df["close"].ewm(span=10, adjust=False).mean()
    df["EMA_50"] = df["close"].ewm(span=50, adjust=False).mean()
    
    # Price Change (%)
    df["price_change"] = df["close"].pct_change() * 100
    
    # Relative Strength Index (RSI)
    delta = df["close"].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    df["RSI_14"] = 100 - (100 / (1 + rs))
    
    # Bollinger Bands
    df["BB_upper"] = df["SMA_10"] + (df["close"].rolling(window=10).std() * 2)
    df["BB_lower"] = df["SMA_10"] - (df["close"].rolling(window=10).std() * 2)
    
    # MACD (Moving Average Convergence Divergence)
    df["MACD"] = df["close"].ewm(span=12, adjust=False).mean() - df["close"].ewm(span=26, adjust=False).mean()
    df["MACD_signal"] = df["MACD"].ewm(span=9, adjust=False).mean()
    
    # Stochastic Oscillator
    df["stoch_k"] = ((df["close"] - df["low"].rolling(window=14).min()) / 
                      (df["high"].rolling(window=14).max() - df["low"].rolling(window=14).min())) * 100
    df["stoch_d"] = df["stoch_k"].rolling(window=3).mean()
    
    # Average True Range (ATR)
    df["TR"] = np.maximum(df["high"] - df["low"], 
                           np.maximum(abs(df["high"] - df["close"].shift()), abs(df["low"] - df["close"].shift())))
    df["ATR"] = df["TR"].rolling(window=14).mean()
    
    # Drop NaN values (from rolling calculations)
    df.dropna(inplace=True)
    
    return df

# Assuming btc_df is your DataFrame with "close" price
btc_df = add_features(btc_df)

# Display first few rows
btc_df.head()
```

```python
import matplotlib.pyplot as plt

# Ensure timestamp is in datetime format
btc_df["timestamp"] = pd.to_datetime(btc_df["timestamp"])

# Compute SMAs if not already computed
btc_df["SMA_10"] = btc_df["close"].rolling(window=10).mean()
btc_df["SMA_50"] = btc_df["close"].rolling(window=50).mean()

# Drop NaNs caused by rolling window
btc_df.dropna(inplace=True)

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(btc_df["timestamp"], btc_df["close"], label="Close Price", color="blue")
plt.plot(btc_df["timestamp"], btc_df["SMA_10"], label="SMA 10", color="red")
plt.plot(btc_df["timestamp"], btc_df["SMA_50"], label="SMA 50", color="green")

plt.title("BTCUSDT Close Price and SMAs")
plt.xlabel("Date")
plt.ylabel("Price (USDT)")
plt.legend(loc="best")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
plt.figure(figsize=(14, 5))
plt.plot(btc_df["timestamp"], btc_df["MACD"], label="MACD", color="blue")
plt.plot(btc_df["timestamp"], btc_df["MACD_signal"], label="Signal Line", color="red")
plt.axhline(0, linestyle="--", color="black")
plt.title("MACD and Signal Line")
plt.xlabel("Date")
plt.ylabel("MACD Value")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 5))
plt.plot(btc_df["timestamp"], btc_df["RSI_14"], label="RSI 14", color="purple")
plt.axhline(70, linestyle="--", color="red", label="Overbought")
plt.axhline(30, linestyle="--", color="green", label="Oversold")
plt.title("Relative Strength Index (RSI)")
plt.xlabel("Date")
plt.ylabel("RSI")
plt.legend()
plt.show()
```

```python
btc_df["signal"] = 0  # No trade
btc_df.loc[btc_df["SMA_10"] > btc_df["SMA_50"], "signal"] = 1  # Buy signal
btc_df.loc[btc_df["SMA_10"] < btc_df["SMA_50"], "signal"] = -1  # Sell signal
```
